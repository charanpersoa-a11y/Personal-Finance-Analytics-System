```python
# fibonacci
import numpy as np
x0=0
x1=1
list_=[]
n=int(input("enter the number of fibonacci sequence :"))
for i in range(n):
    xth=x0+x1
    
    list_.append(x0)

    x0=x1
    x1=xth
even_list=[y for y in list_ if y%2==0]
print(np.sum(even_list))
```

```python
def even(l1):
    c=0
    for val in l1:
        if val%2==0:
            c+=1

    return (c)
# even([1,2,3,4,5,6])
# 3. Find Maximum Number

# Create a function that finds the largest number in a list without using max().

# Input: [5, 2, 9, 1, 7]
# Output: 9
def max_num(list1):
    v=0
    for val in list1:
        if val>v:
            v=val
        
    return v
# max_num([1,5,23,7])
# 4. Reverse a List

# Write a function that returns a reversed list without using reverse().

# Input: [1, 2, 3, 4]
# Output: [4, 3, 2, 1]
def reverse_list(list3):
    for i in range(len(list3)):
        return list3[::-1]
    
    return list3

        
# reverse_list([1,3,4])
# 5. Multiplication Table

# Write a function that prints the multiplication table of a given number up to 10.

# Input: 5

# Output:

# 5 x 1 = 5
# 5 x 2 = 10
# ...
# 5 x 10 = 50
def multi(num):
    for i in range(1,11):
        print(f"{num} x {i} = {num*i}")
        
    
multi(3)
```
