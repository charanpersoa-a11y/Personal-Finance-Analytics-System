Pipeline: filter then transform
Given words = ["hello", "hi", "world", "python", "ok", "generator"], build a two-step generator pipeline:

1. long_words — generator of words longer than 3 characters

2. uppercased — generator that uppercases each word from long_words

```python
words = ["hello", "hi", "world", "python", "ok", "generator"]
long_words=[word for word in words if len(word)>3]
print(long_words)
upper_cased=[word.upper() for word in long_words ]
upper_cased
```

Find the first match efficiently
Given a large range of numbers (1 to 1,000,000), find the first number that is both divisible by 7 and contains the digit "3" in its string representation.

Store it in result. Use next() with a generator — don't compute all matches first.

Expected: 21 ... wait, does 21 contain "3"? Think carefully.

```python
store=(x for x in range(1,1000000) if  x%7==0 and "3" in str(x))
print(next(store))
result = next(x for x in range(1, 1_000_001) if x % 7 == 0 and "3" in str(x))
print(result)
```
