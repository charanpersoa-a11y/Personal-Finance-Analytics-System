```python
import numpy as np

temps = np.array([32, 35, 30, 29, 40, 42, 38, 31, 28, 45])
average_temp=temps.mean()
print("the average temperature is found to be :")
print(average_temp)
above_avg=(temps[temps>average_temp])
print("temperatures above average ")
print(above_avg)
diff=temps-average_temp
print("the difference between temperature and the average temperature")
print(diff)
i=0
streak=0
max_streak=0
for i in range(0,len(temps)):
    
    
    if temps[i]<35:
        streak=0
        continue
    elif temps[i]>35:
        streak+=1
        max_streak=max(max_streak,streak)
    
print("the maximum streak of temperature greater than 35 is :")
print(max_streak)



    

```

Calculate total marks.
Calculate percentage.
Rank students based on percentage.
Find students whose science score is higher than their math score.

```python
import pandas as pd
students = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E"],
    "math": [85, 90, 78, 92, 88],
    "science": [80, 95, 75, 89, 91]
})
print(f"the information of the students data is ")
print(students.head())

students["total_mark"]=students["math"]+students["science"]
print(students["total_mark"])
students["percentage"]=(students["total_mark"]/200)*100
print(students["percentage"])
students=students.sort_values(by="percentage",ascending=False)
print(students)
query=students[students["science"]>students["math"]]
print(query)
```

Calculate total spending per customer.
Find the top spender.
Find customers whose average order value exceeds 250.
Add a column categorizing orders:
Small (<200)
Medium (200–500)
Large (>500)

```python
import pandas as pd
orders = pd.DataFrame({
    "customer": ["A","B","A","C","B","A","D"],
    "amount": [200,500,300,400,150,250,800]
})
orders=orders.groupby('customer')['amount'].sum()
top_spender=orders['amount'].max()
print(top_spender)
print(orders)
```
