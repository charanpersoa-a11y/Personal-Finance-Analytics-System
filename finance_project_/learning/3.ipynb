```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
l3=[]
# for i in range(len(matrix)):
#     for j in range(len(matrix[0])):
#         print(matrix[:j,:i])
def cuboid_coords_list(x, y, z, n):
    return [
        [i, j, k]
        for i in range(x + 1)
        for j in range(y + 1)
        for k in range(z + 1)
        if i + j + k != n
    ]

# Example usage:
x, y, z, n = 1, 1, 1, 2
print(cuboid_coords_list(x, y, z, n))

```
